## Importing Ultralytics

```python
from ultralytics import YOLO
```

## Importing Models

```python
model_n = YOLO("yolo/yolov8n.pt")
```

## Training

```python
model_n.train(data="data.yaml", epochs=100)
metrics = model_n.val()
```

## Testing

```python
model_n = YOLO('runs/detect/train4/weights/best.pt')
```

```python
import os
test = ["dataset/test/images/" + f for f in os.listdir('dataset/test/images')]
```

```python
results = model_n(test)
```

```python
from PIL import Image
images = []
for count, r in enumerate(results):
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    images.append(im)
    im.save('results/' + str(count) + '.jpg')
```

## Displaying some examples

```python
import random
for img in random.sample(images, 10):
    display(img)
```
